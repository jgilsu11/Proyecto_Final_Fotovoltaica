```python
# Tratamiento de datos
# -----------------------------------------------------------------------
import numpy as np
import pandas as pd

# # Para la Visualizacion 
# # -----------------------------------------------------------------------
# import math
# import seaborn as sns
# import matplotlib.pyplot as plt

# # Para los modelos
# # -----------------------------------------------------------------------
# from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# # Importar métricas de evaluación directamente
# from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Para la codificación de las variables categóricas
# -----------------------------------------------------------------------
from sklearn.preprocessing import OneHotEncoder 
from category_encoders import TargetEncoder  

# Para la codificación de las variables numéricas
# -----------------------------------------------------------------------
from sklearn.preprocessing import MinMaxScaler, Normalizer, StandardScaler, RobustScaler

import pickle
import os
import sys 

import warnings
warnings.filterwarnings("ignore")

sys.path.append(os.path.abspath("../src"))   
import soporte_preprocesamiento as f

pd.set_option("display.max_columns", None)
```

```python
df_pequeñas=pd.read_pickle("../datos/red_piso/Datos_formateados/pequeñas_formateadas.pkl")
df_medianas=pd.read_pickle("../datos/red_piso/Datos_formateados/medianas_formateadas.pkl")
df_amplias=pd.read_pickle("../datos/red_piso/Datos_formateados/amplias_formateadas.pkl")
df_grandes=pd.read_pickle("../datos/red_piso/Datos_formateados/grandes_formateadas.pkl")
```

**ABRO LOS PKL DE ENCODING**

```python
frecuencia_diccs = {
    "frecuencia_peq_dicc": "../transformers/preprocesamiento/pequeñas/frecuencia_dicc.pkl",
    "frecuencia_med_dicc": "../transformers/preprocesamiento/medianas/frecuencia_med_dicc.pkl",
    "frecuencia_amp_dicc": "../transformers/preprocesamiento/amplias/frecuencia_amp_dicc.pkl",
    "frecuencia_gra_dicc": "../transformers/preprocesamiento/grandes/frecuencia_gra_dicc.pkl"}

frecuencia_cargada = {nombre: pickle.load(open(ruta, "rb")) for nombre, ruta in frecuencia_diccs.items()}

frecuencia_peq_dicc, frecuencia_med_dicc, frecuencia_amp_dicc, frecuencia_gra_dicc = frecuencia_cargada.values()
```

**ABRO LOS PKL DE ESTANDARIZACIÓN**

```python
scalers = {
    "estandar_peq": "../transformers/preprocesamiento/pequeñas/Standard_scaler.pkl",
    "estandar_med": "../transformers/preprocesamiento/medianas/Minmax_scaler.pkl",
    "estandar_amp": "../transformers/preprocesamiento/amplias/Standard_scaler.pkl",
    "estandar_gra": "../transformers/preprocesamiento/grandes/Robust_scaler.pkl"}

scalers_cargados = {nombre: pickle.load(open(ruta, "rb")) for nombre, ruta in scalers.items()}

estandar_peq, estandar_med, estandar_amp, estandar_gra = scalers_cargados.values()
```

Aplico el Encoding

```python
df_peq_encoded = f.apply_frequency_encoding(df_pequeñas, frecuencia_peq_dicc)
df_med_encoded = f.apply_frequency_encoding(df_medianas, frecuencia_med_dicc)
df_amp_encoded = f.apply_frequency_encoding(df_amplias, frecuencia_amp_dicc)
df_gra_encoded = f.apply_frequency_encoding(df_grandes, frecuencia_gra_dicc)
```

Aplico la Estandarización

PEQUEÑAS

```python
columnas_categoricas=['idq_pv', 'calefac', 'naseos', 'ndormito', 'metrosvi', 'annoedifinterv', 'estadoedif', 'accesoedif', 'ascensor', 'nplazasgar', 'aguacalcent', 'ccaa', 'bonificaciones', "valor"] #incluyo valor aunque no sea categorica para que no la toque
nums_peq = df_peq_encoded.select_dtypes(include=np.number)
cols_a_estandarizar = [col for col in nums_peq.columns if col not in columnas_categoricas]
df_peq_estandarizado = estandar_peq.transform(nums_peq[cols_a_estandarizar])
df_peq_estandarizado = pd.DataFrame(df_peq_estandarizado, columns=cols_a_estandarizar)
df_peq_encoded[cols_a_estandarizar] = df_peq_estandarizado
```

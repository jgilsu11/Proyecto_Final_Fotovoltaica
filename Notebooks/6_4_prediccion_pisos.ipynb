```python
df_pequeñas=pd.read_pickle("../datos/red_piso/Datos_preprocesados/pequeñas_preprocesadas.pkl")
df_medianas=pd.read_pickle("../datos/red_piso/Datos_preprocesados/medianas_preprocesadas.pkl")
df_amplias=pd.read_pickle("../datos/red_piso/Datos_preprocesados/amplias_preprocesadas.pkl")
df_grandes=pd.read_pickle("../datos/red_piso/Datos_preprocesados/grandes_preprocesadas.pkl")
```

```python
modelos = {
    "modelo_peq": "../transformers/modelos/modelo_pequeñas_randomA_1.pkl",
    "modelo_med": "../transformers/modelos/modelo_medianas_randomA_1.pkl",
    "modelo_amp": "../transformers/modelos/modelo_amplias_randomA_1.pkl",
    "modelo_gra": "../transformers/modelos/modelo_grandes_randomA_1.pkl"}

modelos_cargados = {nombre: pickle.load(open(ruta, "rb")) for nombre, ruta in modelos.items()}
modelo_rf_peq, modelo_rf_med, modelo_rf_amp, modelo_rf_gra = modelos_cargados.values()
```

## **VIVIENDAS AMPLIAS**

```python
df_amplias_sin_val=f.predecir_energia_fotovoltaica(df_amplias, modelo_rf_amp, orden_columnas)
```

## **VIVIENDAS GRANDES**

```python
df_grandes_sin_val=f.predecir_energia_fotovoltaica(df_grandes, modelo_rf_gra, orden_columnas)
```

```python
df_grandes_sin_val.head()
```

Separo la de fotovoltaica

```python
df_erfotov_gra=df_grandes_sin_val[["energia_fotovoltaica"]]
df_erfotov_amp=df_amplias_sin_val[["energia_fotovoltaica"]]
df_erfotov_med=df_medianas_sin_val[["energia_fotovoltaica"]]
df_erfotov_peq=df_pequeñas_sin_val[["energia_fotovoltaica"]]
```

Cargo los dataframes antes de codificarlos

```python
df_pequeñas_fin=pd.read_pickle("../datos/red_piso/Datos_formateados/pequeñas_formateadas.pkl")
df_medianas_fin=pd.read_pickle("../datos/red_piso/Datos_formateados/medianas_formateadas.pkl")
df_amplias_fin=pd.read_pickle("../datos/red_piso/Datos_formateados/amplias_formateadas.pkl")
df_grandes_fin=pd.read_pickle("../datos/red_piso/Datos_formateados/grandes_formateadas.pkl")
```

Tras separar las columnas de fotovoltaica de cada uno de ellos para unirsela a cada dataframe antes de preprocesarlo.   

```python
dfs = {
    "Pequeña": (df_pequeñas_fin, df_erfotov_peq),
    "Mediana": (df_medianas_fin, df_erfotov_med),
    "Amplia": (df_amplias_fin, df_erfotov_amp),
    "Grande": (df_grandes_fin, df_erfotov_gra)
}

for categoria, (df, df_erfotov) in dfs.items():
    df["categoria"] = categoria
    dfs[categoria] = pd.concat([df, df_erfotov], axis=1)

df_pequeñas_fin, df_medianas_fin, df_amplias_fin, df_grandes_fin = dfs.values()
```

Unir los distintos data frames (que ya incluyen tanto la columna de valor como la de energía fotovoltaica, además de la de categoría) en un único df.

```python
df_general=pd.DataFrame()
df_general=pd.concat([df_general, df_pequeñas_fin, df_medianas_fin, df_amplias_fin, df_grandes_fin], axis=0)
```

```python
df_general.reset_index(drop=True, inplace=True)
df_general.head()
```
